#  Visualisation

Distribution of Airbnb Booking prices Range using histogram.

```python
plt.figure(figsize=(10,10))
sns.histplot(data=df,x=df['price'],color='orange',bins=50,kde=True)
plt.xlabel('price')
plt.ylabel('Density')
plt.xlim(0,3000)
plt.title('Distribution of Airbnb Booking prices Range')
plt.show()
```

Insights:

Mostly prices are between 0-2000.

Low price is most prefer by people

Find the most busiest Neighbourhood group.

```python
plt.figure(figsize=(10,10))
sns.countplot(data=df,x='neighbourhood_group')
plt.xlabel('Neighbourhood_group')
plt.title('ost busiest Neighbourhood group')
plt.show()
```

Insights:

Most busiest neighbourhood_group is manhattan

Least busiest neighbour_hood group is staten island

What is average price by room type

```python
plt.figure(figsize=(10,10))
sns.barplot(data=df,x='room_type',y='price')
plt.title('Average price by room type')
plt.show()

```

Insights:

The maximum price is entire home/apt and  The minimum price is Shared room
