```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Cars (1).csv')
```

```python
df
```

```python
df.info()
```

```python
sns.barplot(x='cylinders',y='mpg',data=df)
```

```python
a=df['mpg']
b=df['hp']
c=df['time-to-60']
d=df['weightlbs']
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].scatter(a,b)
axs[0, 1].scatter(a,d)
axs[1, 0].scatter(b,a)
axs[1, 1].scatter(b,d)
plt.show()
```

```python
fig=plt.subplots(2,2,figsize=(20,10))
plt.subplot(2,2,1)   #plot-1
sns.scatterplot(x=df['mpg'],y=df['hp'],data=df,hue=df['brand'])
plt.xlabel('MPG')
plt.ylabel('HP')
plt.title('MPG VS HP')

plt.subplot(2,2,2)   #plot-2
sns.scatterplot(x=df['mpg'],y=df['weightlbs'],data=df,hue=df['brand'])
plt.xlabel('MPG')
plt.ylabel('Weight')
plt.title('MPG VS Weight')

plt.subplot(2,2,3)   #plot-3
sns.scatterplot(x=df['hp'],y=df['mpg'],data=df,hue=df['brand'])
plt.xlabel('HP')
plt.ylabel('MPG')
plt.title('HP VS MPG')

plt.subplot(2,2,4)   #plot-4
sns.scatterplot(x=df['hp'],y=df['time-to-60'],data=df,hue=df['brand'])
plt.xlabel('HP')
plt.ylabel('Time-to-60')
plt.title('HP VS Time-to-60')
```

```python
fig=plt.subplots(2,2,figsize=(10,10))
plt.subplot(2,2,1)   #plot-1
sns.barplot(data=df,x='cylinders',y='hp',hue='brand')
plt.xlabel('hp')
plt.ylabel('cylinder')
plt.title('HP VS cylinders')

plt.subplot(2,2,2)   #plot-2
sns.barplot(data=df,x='cylinders',y='mpg',hue='brand')
plt.xlabel('MPG')
plt.ylabel('cylinders')
plt.title('MPG VS cylinders')

plt.subplot(2,2,3)   #plot-3
sns.barplot(data=df,x='cylinders',y='weightlbs',hue='brand')
plt.xlabel('Weight')
plt.ylabel('cylinders')
plt.title('Weight VS cylinders')

plt.subplot(2,2,4)   #plot-4
sns.barplot(data=df,x='cylinders',y='time-to-60',hue='brand')
plt.xlabel('HP')
plt.ylabel('Time-to-60')
plt.title('Time-to-60 VS cylinders')
```

```python
fig=plt.subplots(2,2,figsize=(10,10))
plt.subplot(2,2,1)
sns.boxplot(data=df,x='cylinders',y='hp',hue='brand')
plt.xlabel('Cylinders')
plt.ylabel('HP')

plt.subplot(2,2,2)
sns.boxplot(data=df,x='cylinders',y='mpg',hue='brand')
plt.xlabel('Cylinders')
plt.ylabel('MPG')

plt.subplot(2,2,3)
sns.boxplot(data=df,x='weightlbs',y='cylinders',hue='brand')
plt.xlabel('weight')
plt.ylabel('cylinder')

plt.subplot(2,2,4)
sns.boxplot(data=df,x='cylinders',y='time-to-60',hue='brand')
plt.xlabel('Cylinders')
plt.ylabel('Time-to-60')
```
