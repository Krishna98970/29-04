Create scatter plot for the following data,make sure all the plots appear in the same plane with x-labels and y-labels evenly spaced.

(mpg-hp,mpg-weight) and (hp-mpg,hp-time-to-60) with respect to brand.

```python
fig=plt.subplots(2,2,figsize=(20,10))
plt.subplot(2,2,1)   #plot-1
sns.scatterplot(x=df['mpg'],y=df['hp'],data=df,hue=df['brand'])
plt.xlabel('MPG')
plt.ylabel('HP')
plt.title('MPG VS HP')

plt.subplot(2,2,2)   #plot-2
sns.scatterplot(x=df['mpg'],y=df['weightlbs'],data=df,hue=df['brand'])
plt.xlabel('MPG')
plt.ylabel('Weight')
plt.title('MPG VS Weight')

plt.subplot(2,2,3)   #plot-3
sns.scatterplot(x=df['hp'],y=df['mpg'],data=df,hue=df['brand'])
plt.xlabel('HP')
plt.ylabel('MPG')
plt.title('HP VS MPG')

plt.subplot(2,2,4)   #plot-4
sns.scatterplot(x=df['hp'],y=df['time-to-60'],data=df,hue=df['brand'])
plt.xlabel('HP')
plt.ylabel('Time-to-60')
plt.title('HP VS Time-to-60')
```

Create barplots for the following ,make sure all the plots appear in the same plane with xlabel and y label evenly spaced.

hp,mpg,weight,time-to-60 with respect to brand and to the number of cylinders.

hp vs cylinders

```python
fig=plt.subplots(2,2,figsize=(10,10))
plt.subplot(2,2,1)
sns.barplot(data=df,x='cylinders',y='hp',hue='brand')
plt.xlabel('Cylinders')
plt.ylabel('HP')

plt.subplot(2,2,2)
sns.barplot(data=df,x='cylinders',y='mpg',hue='brand')
plt.xlabel('Cylinders')
plt.ylabel('MPG')

plt.subplot(2,2,3)
sns.barplot(data=df,x='cylinders',y='weightlbs',hue='brand')
plt.xlabel('Cylinders')
plt.ylabel('Weight')

plt.subplot(2,2,4)
sns.barplot(data=df,x='cylinders',y='time-to-60',hue='brand')
plt.xlabel('Cylinders')
plt.ylabel('Time-to-60')
```

Create a box-plot for the following with repect to brand:

```python
fig = plt.subplots(2,2,figsize = (20,10))
plt.subplot(2,2,1)
sns.boxplot(x='cylinders', y='hp', data = df, hue = df.brand)
plt.xlabel('HP')
plt.ylabel('CYLENDERS')
plt.title('HP VS CYLENDERS')
plt.subplot(2,2,2)
sns.boxplot(x=df.cylinders, y=df.mpg, data = df, hue = df.brand)
plt.xlabel('MPG')
plt.ylabel('CYLENDERS')
plt.title('MPG VS CYLENDERS')
plt.subplot(2,2,3)
sns.boxplot(x=df.cylinders, y=df.weightlbs, data = df, hue = df.brand)
plt.xlabel('WEIGHT')
plt.ylabel('CYLENDERS')
plt.title('WEIGHT VS CYLENDERS')
plt.subplot(2,2,4)
sns.boxplot(x=df.cylinders, y=df['time-to-60'], data = df, hue = df.brand)
plt.xlabel('TIME-TO-60')
plt.ylabel('CYLENDERS')
plt.title('TIME-TO-60 VS CYLENDERS')
plt.show()
```

Create a histogram for the following:
hp,mpg,time-to-60,weight,cubicinches,year with respect to brand.Make sure to plot all the plots in the same plane with x and y label evenly spaced.

```python
fig=plt.subplots(3,2,figsize=(20,20))
plt.subplot(3,2,1)
sns.histplot(data=df,x='hp',hue='brand')
plt.xlabel('HP')

plt.subplot(3,2,2)
sns.histplot(data=df,x='mpg',hue='brand')
plt.xlabel('MPG')

plt.subplot(3,2,3)
sns.histplot(data=df,x='time-to-60',hue='brand')
plt.xlabel('Time-to-60')

plt.subplot(3,2,4)
sns.histplot(data=df,x='weightlbs',hue='brand')
plt.xlabel('Weight')

plt.subplot(3,2,5)
sns.histplot(data=df,x='cubicinches',hue='brand')
plt.xlabel('Cubicinches')

plt.subplot(3,2,6)
sns.histplot(data=df,x='year',hue='brand')
plt.xlabel('Year')
plt.show()
```

Create a heatmap for the entire data to study the correlation between each of the columns.
